## KNN(n_neighbors=100)

```python
import pandas as pd
import numpy as np

x3_train,x3_test,y3_train,y3_test=train_test_split(x,y,test_size=0.25,random_state=0)

# fonction pour KNN
def KNN(k):
    cl = KNeighborsClassifier(n_neighbors=k)
    cl.fit(x3_train, y3_train)
    y3_pred = cl.predict(x3_test)
    y3_pred_proba = cl.predict_proba(x3_test)
    #calcul des metrics
    score3_train = cl.score(x3_train,y3_train)
    score3_test = cl.score(x3_test,y3_test)
    precision3 = precision_score(y3_test, y3_pred, average='macro')
    recall3 = recall_score(y3_test, y3_pred, average='macro')
    accuracy3 = accuracy_score(y3_test, y3_pred)
    f1_score3_B = f1_score(y3_test, y3_pred, pos_label=0)
    f1_score3_M = f1_score(y3_test, y3_pred, pos_label=1)
    ar = np.array([[score3_train, score3_test, precision3,recall3,accuracy3,f1_score3_B,f1_score3_M]])
    df = pd.DataFrame(ar,index = ['KNN(K={})'.format(k)],columns=['score_train','score_test','precision','recall','accuracy','f1_score_B','f1_score_M'])
    
    #ROC
    logit_roc_auc3 = roc_auc_score(y3_test, y3_pred)
    fpr3, tpr3, thresholds3 = roc_curve(y3_test, y3_pred_proba[:,1])
    plt.figure()
    plt.plot(fpr3, tpr3, label='Logistic Regression (area = %0.2f)' % logit_roc_auc3)
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic with KNN(k={})'.format(k))
    plt.legend(loc="lower right")
    plt.savefig('Log_ROC')
    plt.show()
    return df
```

# Cross Validation sans standardScaler

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

#Fractionnement des donnÃ©es
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)

K = [5,3,10,100,200]
def cross_val(K):
    val_score = []
    for k in K:
        score_val = cross_val_score(KNeighborsClassifier(k), x_train, y_train, cv=5).mean()
        val_score.append(score_val)
    val_mean = (np.array(val_score)).mean()
    val_std = (np.array(val_score)).std()
    #print(val_mean)
    plt.plot(val_score)
    plt.title('Cross validation with KNN (k = {})'.format(K))
    return 'val_score_mean = {}'.format(val_mean),  'val_score_std = {}'.format(val_std)

cross_val(K)
```

# KNN avec standardScaler 

## Pipeline

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
import numpy as np



def pipe_SC_KNN(k):

    #fractionner les donnÃ©es
    X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25,random_state=0)
    
    #choisir le model et le preprocessing
    pipe = Pipeline([('scaler', StandardScaler()), ('KNN', KNeighborsClassifier(n_neighbors=k))])
    pipe.fit(X_train, Y_train)
    
    #calcul y_pred et y_pred_proba
    Y_pred = pipe.predict(X_test)
    Y_pred_proba = pipe.predict_proba(X_test)
    
    #culcule des metrics
    score_test = pipe.score(X_test, Y_test)
    score_train = pipe.score(X_train, Y_train)
    precision = precision_score(Y_test, Y_pred, average='macro')
    recall = recall_score(Y_test, Y_pred, average='macro')
    accuracy = accuracy_score(Y_test, Y_pred)
    f1_score_B = f1_score(Y_test, Y_pred, pos_label=0)
    f1_score_M = f1_score(Y_test, Y_pred, pos_label=1)
    resultats = [score_train,score_test,precision,recall,accuracy,f1_score_B,f1_score_M]
    
    #matrice de confusion
    print("matrice de confusion KNN(k={}): \n".format(k),confusion_matrix(Y_test,Y_pred))
    
    #creer le dataframe pour afficher un tableau
    ar = np.array([resultats])
    DF = pd.DataFrame(ar, index = ['KNN + Scaler + K = {}'.format(k)],
            columns=['score_train','score_test','precision','recall','accuracy','f1_score_B','f1_score_M'])
    
    #ROC
    logit_roc_auc = roc_auc_score(Y_test, pipe.predict(X_test))
    fpr, tpr, thresholds = roc_curve(Y_test, pipe.predict_proba(X_test)[:,1])
    plt.figure()
    plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic with KNN (k = {})'.format(k))
    plt.legend(loc="lower right")
    plt.savefig('Log_ROC')
    plt.show()
    
    return DF
```

# Cross validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline


#Fractionnement des donnÃ©es
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)

K = [5,3,10,100,200]
def cross_val_scaler(K):
    val_score = []
    for k in K:
        pipe = Pipeline([('scaler', StandardScaler()), ('KNN', KNeighborsClassifier(n_neighbors=k))])
        pipe.fit(x_train, y_train)
        score_val = cross_val_score(pipe, x_train, y_train, cv=5,  scoring='accuracy').mean()
        val_score.append(score_val)
    val_mean = (np.array(val_score)).mean()
    val_std = (np.array(val_score)).std()
    #print(val_mean)
    plt.plot(val_score)
    plt.title('Cross validation with KNN (k = {})'.format(K))
    return 'val_score_mean = {}'.format(val_mean),  'val_score_std = {}'.format(val_std)

cross_val_scaler(K)
```

On remarque que pour les k = [5,3,10] le model a une performace (accuracy) de 96% par contre elle diminue pour k=[100,200] 

```python
K = range(2,50)
cross_val_scaler(K)
```

Ici on remarque que le k obtimal est [8,9]  pour des donnÃ©es standardisÃ©es
ce n'est pas le mÃªme avec les donnÃ©es non standardisÃ©es
